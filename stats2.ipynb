In probabilitry theory , a long normal distribution is a continuous probability distribution of a random variable whose logarithm is normal distributed . thus, if the random variable x is log normally distributed , then y = in(x) has a normal distribution . equivalently , if y has a nnormal distribution , then the exponential function of y,x = exp(y), has a log normal distribution .

```python
import matplotlib.pyplot as plt 
import scipy.stats as stat
import pylab

def plot_data(data):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)

    sns.histplot(data,kde=True)
    plt.subplot(1,2,2)

    stat.probplot(data,dist="norm",plot=pylab)   # qq plot quantile quantile plot ( Bplot between theortical and real time data or assumption )
    plt.show()
```

why we use qq plot 

to check whether my data is normlaly distributed or not 

```python
plot_data(lognormal_distributed_data)
```

```python
plot_data(np.log(lognormal_distributed_data))
```

CENTRAL LIMIT THEOREM

Suppose you have a population with any shape of probability distriibution ( not necessary normal) , and you draw random samples **(at least 30)** of a fixed size from this population . As the sample size increases **(it might be anything and take as much as data)** , the distribution of the sample means approaches distribution , regardless of the shape of the original population distribution .

```python
population=np.random.randint(10,20,50)
```

```python
population
```

```python
np.mean(population)
```

```python
np.median(population)
```

```python
from statistics import mode
mode(population)
```

```python
# random sample from the population 
sample=np.random.choice(population,20)
```

```python
sample
```

```python
print(f"mean of the sample {np.mean(sample)}")
print(f"median of the sample {np.median(sample)}")
print(f"mode of the sample {mode(sample)}")
```

```python
sample1=np.random.choice(population,20)
sample2=np.random.choice(population,20)
sample3=np.random.choice(population,20)
sample4=np.random.choice(population,20)
```

```python
print(sample1)
print(60*"=")
print(sample2)
print(60*"=")
print(sample3)
print(60*"=")
print(sample4)
print(60*"=")
```

```python
mean_of_the_sample=[]

all_sample=[sample1,sample2,sample3,sample4]

for sample in all_sample:
    print(np.mean(sample))
    mean_of_the_sample.append(np.mean(sample))
```
